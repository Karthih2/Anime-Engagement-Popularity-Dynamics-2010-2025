```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Loading Raw Data").getOrCreate()

bronze_df = spark.read.csv(
    "/Volumes/workspace/default/anime_raw/anime.csv",
    header=True,
    inferSchema=True,
    multiLine=True,        # ← KEY FIX: handles newlines inside fields
    quote='"',             # ← treats quoted fields correctly
    escape='"',            # ← handles escaped quotes inside fields
    encoding="UTF-8"       # ← handles special characters (anime titles!)
)

display(bronze_df)
```

```python
bronze_df.write.mode("overwrite").saveAsTable("bronze_anime")
```
